```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```python
df = spark.read.csv('./data/appl_stock.csv', inferSchema = True, header = True)
```

Spark can infer the schema of the data provided and parse the headers from a csv.

- If the inferSchema parameter is set to False then the content is read as string. 
- If the header is set to false then the first row will also be included in the dataframe.

#### Let us check the schema of the dataframe 

```python
df.printSchema()
```

## Filtering Data

We have the ability to filter out data using the conditions. Spark can handle SQL commands to filter out data or using the Dataframe methods. 

### SQL Approach


```python
# Using SQL

df.filter("Close<500").show()
```

```python
# SQL Approach with Specific column
# Using .select() method 

df.filter("Close < 500").select('Open').show()
```

```python
# SQL Approach with Specific columns - note column names wrapped in array now 
# Using .select() method 

df.filter("Close < 500").select(['Open', 'Close']).show()
```

#### Multiple Conditions

- and operation

```python
# SQL Approach with Specific columns - note column names wrapped in array now 
# Using .select() method 

df.filter("Close < 500 and Open > 200").select(['Open', 'Close']).show()
```

- or operation

```python
df.filter("Close < 200 or Open > 200").show()
```

- not operation

```python
df.filter("Close < 200 or not Open > 200").show()
```

- equal to operation

```python
df.filter("Low = 197.16").show()
```

### Using Python operators

While using python operators, it is important to call the entire coulumn using the dataframe using the format df["column name"]

```python
# Using the python comparison operator

df.filter( df["Close"] < 500).select(['Open', 'Close']).show()
```

#### Multiple conditions 

 - and operation => & Operator 

```python
# Important to separate our conditions using paranthesis

df.filter( (df["Close"] < 500) & (df["Open"] > 200) ).select(['Open', 'Close']).show()
```

- or operation => | operator 

```python
df.filter( (df["Close"] < 200) | (df['Open'] > 200) ).show()
```

- not operation => ~ operator 

```python
df.filter( (df["Close"] < 200) | ~(df['Open'] > 200) ).show()
```

- equals operation => == operator 

```python
df.filter( df["Low"] ==197.16 ).show()
```
